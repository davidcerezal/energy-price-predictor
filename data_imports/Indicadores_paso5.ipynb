```python
df_temp_hour = df
df_temp_hour.head()
```

```python
from datetime import datetime
df_temp_hour['date'] = pd.to_datetime(df_temp_hour.index)
df_temp_hour['date_timestamp'] = df_temp_hour['date'].apply(lambda x:  datetime.timestamp(x))
df_temp_hour.head()
```

```python
data_total = pd.read_csv("data_total.csv") 
data_total.head()
```

```python
total_df = pd.merge(data_total, df_temp_hour, on='date_timestamp')
total_df.tail()
```
