## Creación de la variable target

```python
def add_target_tomorrow(df):
    for i, row in df.iterrows():
        if (i+24 >= 0 and i+24 <= 30554):
            df.set_value(i,'PVPC-target',df.loc[i+24, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-target',0) 

add_target_tomorrow(data_total)
data_total.to_csv('data_total.csv')
```

## Creación de variantes del dataset. 

Algunos modelos no secuenciales necesitan en todo momento el precio de dias, horas u semanas anteriores. Para ello vamos a crear 4 variables opcionales con el precio 1h, 2h, 1d, 1w respectivamente. 

```python
def filter_array_for_non_serial(df):
    for i, row in df.iterrows():
        if (i-1 >= 0):
            df.set_value(i,'PVPC-1h',df.loc[i-1, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-1h',0) 
        
        if (i-2 >= 0):
            df.set_value(i,'PVPC-2h',df.loc[i-2, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-2h',0) 
        
        if (i-24 >= 0):
            df.set_value(i,'PVPC-1d',df.loc[i-24, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-1d',0) 
            
        if (i-168 >= 0):
            df.set_value(i,'PVPC-1w',df.loc[i-168, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-1w',0)  

filter_array_for_non_serial(data_total_dropped)
```
