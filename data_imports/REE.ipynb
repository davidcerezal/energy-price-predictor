Generación (o actualización) del archivo local de datos de PVPC recogidos en https://www.esios.ree.es/

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from esiosdata import PVPC
from esiosdata.prettyprinting import *
import pandas as pd

from esiosdata import DatosREE
ree_handler = DatosREE()
df_ree = ree_handler.data['data']
```

#### Información sobre el archivo local de datos de PVPC:

```python
df_ree.tail()
```

```python
df_temp = df_ree
print(df_ree.shape)
```

```python
df_temp['date'] = pd.to_datetime(df_temp.index)
df_temp_day = df_temp.groupby(pd.Grouper(key='date', freq='1d')).mean().dropna()
df_temp_month = df_temp.groupby(pd.Grouper(key='date', freq='1M')).mean().dropna()

df_temp_day.loc[:, 'Total']= df_temp_day.sum(axis=1)
df_temp_month.loc[:, 'Total']= df_temp_month.sum(axis=1)

df_temp_day['date'] = df_temp_day.index
df_temp_month['date'] = df_temp_month.index
print(df_temp_day.shape)
print(df_temp_month.shape)
df_temp_month.tail()
```

```python
import seaborn as sns
scatter, ax = plt.subplots()

ax = sns.regplot(x='date', y="Total", data=df_temp_month, color="orange")
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

```python
scatter, ax = plt.subplots()

ax = sns.regplot(x='date', y="Total", data=df_temp_day, color="orange")
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```
