```python
data_total = pd.read_csv("data_total.csv") 
data_total.head()
```

```python
print(data_total.shape)
print(data_total.columns)
```

```python
data_total_dropped = data_total.drop_duplicates(subset=['date_timestamp'])
data_total_dropped.head(1300)
```

```python
print(data_total_dropped.shape)
print(data_total_dropped.columns)
```

```python
data_total_dropped = data_total_dropped.loc[:, ~data_total_dropped.columns.str.contains('^Unnamed')]
data_total_dropped.to_csv('data_total.csv')
```

```python
def filter_array_for_non_serial(df):
    for i, row in df.iterrows():
        if (i-1 >= 0):
            df.set_value(i,'PVPC-1h',df.loc[i-1, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-1h',0) 
        
        if (i-2 >= 0):
            df.set_value(i,'PVPC-2h',df.loc[i-2, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-2h',0) 
        
        if (i-21 >= 0):
            df.set_value(i,'PVPC-1d',df.loc[i-21, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-1d',0) 
            
        if (i-147 >= 0):
            df.set_value(i,'PVPC-1w',df.loc[i-147, 'PVPC_DEF']) 
        else:
            df.set_value(i,'PVPC-1w',0) 

filter_array_for_non_serial(data_total_dropped)
```
