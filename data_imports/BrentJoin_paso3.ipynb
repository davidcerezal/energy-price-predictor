## Obtención de datos BRENT

Obtención y unión del dataset existente con los datos del precio del barril de brent. 

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from esiosdata import PVPC
from esiosdata.prettyprinting import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
data_brent = pd.read_csv("brent-daily.csv") 
data_brent = data_brent[data_brent['Date']>='2015-10-01']
```

```python
data_brent.rename({'Date': 'date_day'}, axis=1, inplace=True)
data_brent.head()
```

## Merge con los df anteriores

```python
data_precios = pd.read_csv("data_precios.csv") 
data_consumos = pd.read_csv("data_demanda_generacion.csv") 
data_consumos.head()
```

```python
total_df = pd.merge(data_precios, data_consumos, on='date_timestamp')
total_df.head(100)
```

```python
total_df = pd.merge(total_df, data_brent, on='date_day', how='left')
total_df.rename({'Price': 'Brent_price'}, axis=1, inplace=True)
total_df.head(100)
```

```python
total_df[['Brent_price']] = total_df[['Brent_price']].fillna(method='ffill')
total_df.head(100)
```

```python
total_df.to_csv('data_total.csv')
```
