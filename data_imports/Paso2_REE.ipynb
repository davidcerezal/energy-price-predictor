# Paso 2.- Datos de REE

## Obtención de datos REE

Obtención de los datos de Consumo y generación, utilizando la misma librería de EsisoData. 

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from esiosdata import PVPC
from esiosdata.prettyprinting import *
import pandas as pd
from datetime import datetime

from esiosdata import DatosREE
ree_handler = DatosREE()
```

#### Información sobre el archivo local de datos de Consumo

```python
df_ree = ree_handler.data['data']
df_ree.head()
```

```python
df_ree['date'] = pd.to_datetime(df_ree.index)
df_ree = df_ree.groupby(pd.Grouper(key='date', freq='1h')).mean().dropna()

df_ree['date'] = pd.to_datetime(df_ree.index)
df_ree['date_timestamp'] = df_ree['date'].apply(lambda x:  datetime.timestamp(x))
df_ree['date_day'] = df_ree['date'].apply(lambda x: datetime.strftime(x, '%Y-%m-%d'))
print(df_ree.shape)
```

```python
df_ree = df_ree[df_ree['date'] >= '2015-10-01']
df_ree.head()
```

```python
df_ree = df_ree.loc[:,('dem', 'eol', 'nuc', 'sol', 'solFot', 'solTer', 'termRenov', 'pre', 'pro', 'date', 'date_timestamp', 'date_day')]
df_ree.rename({'dem': 'Demanda', 'eol': 'Eolica', 'nuc': 'Nuclear', 'sol':'Solar'}, axis=1, inplace=True)
df_ree.rename({'solFot': 'Solar_Fotovoltaica', 'solTer': 'Solar_Termica', 'termRenov': 'Termica_Renovable'}, axis=1, inplace=True)
df_ree.rename({'pre': 'Prevista', 'pro': 'Programada'}, axis=1, inplace=True)

df_ree.tail()
```

```python
df_ree.to_csv('data_demanda_generacion.csv')
```
